```python
# üìò BELAJAR DASAR PYTHON: Variabel, f-string, dan Tipe Data
# -----------------------------------------------------------

# üéØ TUJUAN:
# 1. Memahami variabel dan tipe data dasar: String, Integer, Float, Boolean
# 2. Menggunakan f-string untuk mencetak output dengan variabel
# 3. Mengetahui blindspot (kesalahan umum) dan tips belajar Python

# üß† APA ITU VARIABEL?
# Variabel adalah "wadah" untuk menyimpan nilai.
# Seolah-olah, variabel itu adalah nilai yang dikandungnya.

# Contoh:
first_name = "Zahid"   # String (teks)
food = "NasGor"        # String

# üî§ STRINGS (teks)
# String adalah data yang berisi teks, selalu diapit tanda kutip (" " atau ' ')
# Kita bisa menampilkan variabel string dalam kalimat menggunakan f-string.

print(f"Hello {first_name}")          # Output: Hello Zahid
print(f"Do you like {food}?")         # Output: Do you like NasGor?

# ‚ö†Ô∏è BLINDSPOT:
# Salah tulis nama variabel (typo) bikin error!
# Contoh salah: print(f"Hello {fisrt_name}") ‚ùå

# -----------------------------------------------------------
# üî¢ INTEGERS (bilangan bulat)
age = 24              # Integer
quantity = 3          # Integer

print(f"My age is {age}")                      # Output: My age is 24
print(f"You are buying {quantity} {food}")     # Output: You are buying 3 NasGor

# ‚ö†Ô∏è BLINDSPOT:
# Jangan tambahkan kutip pada angka, karena akan jadi string!
# Salah: age = "24" (kalau mau hitung usia nanti bisa error)

# -----------------------------------------------------------
# üî£ FLOAT (bilangan desimal)
price = 10.99
gpa = 3.48
distance = 2.5

print(f"The price is ${price}")           # Output: The price is $10.99
print(f"My GPA is {gpa}")                 # Output: My GPA is 3.48
print(f"You ran {distance} kilometers")   # Output: You ran 2.5 kilometers

# ‚ö†Ô∏è BLINDSPOT:
# Float tidak cocok untuk jumlah utuh seperti jumlah barang (gunakan integer)

# -----------------------------------------------------------
# ‚úÖ BOOLEAN (benar/salah)
is_student = False
for_sale = True
is_online = False

# Boolean sering dipakai dalam kondisi IF
# IF akan berjalan jika nilai boolean = True

if is_student:
    print("You are a student")
else:
    print("You are NOT a student")

if for_sale:
    print("This item is for sale")
else:
    print("This item is NOT for sale")

if is_online:
    print("This course is ONLINE")
else:
    print("This course is OFFLINE")

# ‚ö†Ô∏è BLINDSPOT:
# True dan False harus diawali huruf kapital
# Salah: true, false ‚Üí akan error karena tidak dikenali Python

# -----------------------------------------------------------
# üì¶ BONUS: APA ITU F-STRING?
# f-string adalah cara praktis untuk menyisipkan variabel ke dalam string.

# Contoh:
name = "Fauzan"
print(f"My name is {name}")  # f sebelum tanda kutip penting!
# Tanpa f ‚Üí variabel tidak akan diproses

# Salah: print("My name is {name}") ‚ùå ‚Üí akan mencetak: My name is {name}
# Benar: print(f"My name is {name}") ‚úÖ

# -----------------------------------------------------------
# üìå KESIMPULAN SINGKAT:

# | Tipe    | Contoh Nilai    | Tipe Data | Kegunaan                         |
# |---------|------------------|-----------|----------------------------------|
# | str     | "Halo"           | Teks      | Nama, alamat, teks umum          |
# | int     | 25               | Angka     | Usia, jumlah barang              |
# | float   | 3.14             | Desimal   | Harga, jarak, nilai rata-rata    |
# | bool    | True / False     | Logika    | Kondisi, status, flag            |

# -----------------------------------------------------------
# ‚úÖ Selesai! Kamu baru saja mempelajari dasar Python yang sangat penting!
# üöÄ Silakan modifikasi nilai dan bereksperimen untuk belajar lebih dalam.
```

English Language

```python
# üìò PYTHON BASICS: Variables, f-strings, and Data Types
# ------------------------------------------------------

# üéØ OBJECTIVES:
# 1. Understand variables and basic data types: String, Integer, Float, Boolean
# 2. Learn how to use f-strings to print variables inside text
# 3. Recognize common blindspots (mistakes beginners often make)

# üß† WHAT IS A VARIABLE?
# A variable is a "container" that holds a value.
# Think of it as if the variable *is* the value it stores.

# Example:
first_name = "Zahid"   # String (text)
food = "Fried Rice"     # String

# üî§ STRINGS (Text)
# A string is a piece of text inside quotes (" " or ' ')
# We use f-strings to include variables inside printed text.

print(f"Hello {first_name}")            # Output: Hello Zahid
print(f"Do you like {food}?")           # Output: Do you like Fried Rice?

# ‚ö†Ô∏è BLINDSPOT:
# Typing mistakes (typos) in variable names will cause errors!
# Incorrect: print(f"Hello {fisrt_name}") ‚ùå

# ------------------------------------------------------
# üî¢ INTEGERS (Whole numbers)
age = 24             # Integer
quantity = 3         # Integer

print(f"My age is {age}")                      # Output: My age is 24
print(f"You are buying {quantity} {food}")     # Output: You are buying 3 Fried Rice

# ‚ö†Ô∏è BLINDSPOT:
# Don‚Äôt use quotes for numbers, or Python will treat them as strings!
# Incorrect: age = "24"

# ------------------------------------------------------
# üî£ FLOAT (Decimal numbers)
price = 10.99
gpa = 3.48
distance = 2.5

print(f"The price is ${price}")           # Output: The price is $10.99
print(f"My GPA is {gpa}")                 # Output: My GPA is 3.48
print(f"You ran {distance} kilometers")   # Output: You ran 2.5 kilometers

# ‚ö†Ô∏è BLINDSPOT:
# Floats are not ideal for counting items (use integers instead)

# ------------------------------------------------------
# ‚úÖ BOOLEAN (True or False)
is_student = False
for_sale = True
is_online = False

# Boolean values are often used in IF conditions
# IF runs the code only if the condition is True

if is_student:
    print("You are a student")
else:
    print("You are NOT a student")

if for_sale:
    print("This item is for sale")
else:
    print("This item is NOT for sale")

if is_online:
    print("This course is ONLINE")
else:
    print("This course is OFFLINE")

# ‚ö†Ô∏è BLINDSPOT:
# True and False must be capitalized in Python
# Incorrect: true, false ‚ùå

# ------------------------------------------------------
# üì¶ BONUS: WHAT IS AN F-STRING?
# An f-string lets you insert variables directly inside strings.

# Example:
name = "Fauzan"
print(f"My name is {name}")  # The 'f' before the string is important!

# Without the 'f' ‚Üí the variable won't be processed
# Incorrect: print("My name is {name}") ‚ùå
# Correct:   print(f"My name is {name}") ‚úÖ

# ------------------------------------------------------
# üìå QUICK REFERENCE:

# | Type   | Example Value     | Data Type | Use Case                       |
# |--------|-------------------|-----------|--------------------------------|
# | str    | "Hello"           | Text      | Names, messages, text values   |
# | int    | 25                | Number    | Age, item count                |
# | float  | 3.14              | Decimal   | Prices, scores, distances      |
# | bool   | True / False      | Logic     | Conditions, states, switches   |

# ------------------------------------------------------
# ‚úÖ DONE! You've just learned the most important Python basics.
# üöÄ Try changing values and experimenting to go even further!
```
